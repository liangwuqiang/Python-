# 02 基础

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 散点图

```python
x = np.random.rand(20)
y = np.random.rand(20)
size = np.random.rand(20) * 200
color = np.random.rand(20)
plt.scatter(x, y, size, color)

plt.colorbar()  # 显示颜色条
plt.show()
```

### 多图

```python
x = np.linspace(0, 2 * np.pi, 50)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure()
plt.plot(y_sin)
plt.figure()  # 产生新的图像
plt.plot(y_cos)
plt.show()
```

```python
plt.figure()
plt.subplot(1, 2, 1)  # subplot(row, column, index)
plt.plot(y_sin)
plt.subplot(1, 2, 2)
plt.plot(y_cos)
plt.show()
```

### 拼图

```python
plt.plot(y_sin)
plt.plot(y_cos)
plt.show()
```

### 图例

```python
l_sin, = plt.plot(y_sin)
l_cos, = plt.plot(y_cos)
plt.legend((l_cos, l_sin), ('cos', 'sin'))
plt.show()
```

### 坐标轴、标题、网格

```python
plt.plot(x, np.sin(x))
plt.xlabel('radians')
plt.ylabel('amplitude', fontsize='large')
plt.title('Sin(x)')
plt.grid()
plt.show()
```

## imshow 显示图片

```python
from scipy.misc import ascent
# img = lena()  # please use ascent() or face() instead
img = ascent()
img
```

```python
plt.imshow(img,
      extent = [-25, 25, -25, 25],  # 设置坐标范围
      cmap = plt.cm.bone)  # 设置colormap
plt.colorbar()
plt.show()
```

```python
dir(plt.cm)  # 查看colormap的种类
```

```python
plt.imshow(img, cmap=plt.cm.RdGy_r)
plt.show()
```
