# 09 应用实例

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi)
y_cos, y_sin = np.cos(x), np.sin(x)
```

### 默认显示

```python
plt.plot(x, y_cos)
plt.plot(x, y_sin)
plt.show()
```

### 图形修饰

```python
plt.figure(figsize=(10,6), dpi=80)  # 设置图像大小 （单位英寸）

plt.plot(x, y_cos, 'b-', lw=2.5)  # 指定颜色、线性、线宽
plt.plot(x, y_sin, 'r-', lw=2.5)

plt.xlim(x.min() * 1.1, x.max() * 1.1)  # 控制显示区域
plt.ylim(y_cos.min() * 1.1, y_cos.max() * 1.1)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],  # 设置刻度
           ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'], fontsize ='x-large')
plt.yticks([-1, 0, 1], 
           ['$-1$', '$0$', '$+1$'], fontsize ='x-large')

plt.show()
```

### 改变坐标轴位置

```python
plt.figure(figsize=(10,6), dpi=80)  # 设置图像大小 （单位英寸）

plt.plot(x, y_cos, 'b-', lw=2.5)  # 指定颜色、线性、线宽
plt.plot(x, y_sin, 'r-', lw=2.5)

plt.xlim(x.min() * 1.1, x.max() * 1.1)  # 控制显示区域
plt.ylim(y_cos.min() * 1.1, y_cos.max() * 1.1)

# -----------------------------------------
ax = plt.gca()  # 得到轴的句柄

ax.spines['right'].set_color('none')  # ax.spines参数表示四个坐标轴线 将右边和上边的颜色设为透明
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')  # 将 x 轴的刻度设置在下面的坐标轴上
ax.spines['bottom'].set_position(('data',0))  # 设置x轴位置
ax.yaxis.set_ticks_position('left')  # 将 y 轴的刻度设置在左边的坐标轴上
ax.spines['left'].set_position(('data',0))  # 设置y轴位置
# -----------------------------------------

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],   # 设置刻度
           ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'], size ='x-large')
plt.yticks([-1, 1], ['$-1$', '$+1$'], size ='x-large')

plt.show()
```

### 注释

```python
plt.figure(figsize=(10,6), dpi=80)  # 设置图像大小 （单位英寸）

plt.plot(x, y_cos, 'b-', lw=2.5)  # 指定颜色、线性、线宽
plt.plot(x, y_sin, 'r-', lw=2.5)

plt.xlim(x.min() * 1.1, x.max() * 1.1)  # 控制显示区域
plt.ylim(y_cos.min() * 1.1, y_cos.max() * 1.1)

ax = plt.gca()  # 得到轴的句柄

ax.spines['right'].set_color('none')  # ax.spines参数表示四个坐标轴线 将右边和上边的颜色设为透明
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')  # 将 x 轴的刻度设置在下面的坐标轴上
ax.spines['bottom'].set_position(('data',0))  # 设置x轴位置
ax.yaxis.set_ticks_position('left')  # 将 y 轴的刻度设置在左边的坐标轴上
ax.spines['left'].set_position(('data',0))  # 设置y轴位置

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],   # 设置刻度
           ['$-\pi$', '$-\pi/2$', '$0$', '$\pi/2$', '$\pi$'], size ='x-large')
# plt.yticks([-1, 0, 1], ['$-1$', '$0$', '$+1$'], fontsize ='x-large')
plt.yticks([-1, 1], ['$-1$', '$+1$'], size ='x-large')

# ----------------------------------------- 添加图例
l = plt.legend(['cosine', 'sine'], loc='upper left', frameon=False)  # frameon表示图例周围是否需要边框

# ----------------------------------------- 注释
x = 2 * np.pi / 3  # 数据点

plt.plot([x, x],[0, np.sin(x)], 'r--', lw=2.5)  # 红色虚线
plt.scatter([x, ],[np.sin(x), ], s=50, c='red')  # 红点

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', # 文本 值是怎么得到的？
             xy=(x, np.sin(x)), # 数据点坐标位置
             xycoords='data',   # 坐标相对于数据
             xytext=(+10, +30), # 文本位置坐标 相对位置
             textcoords='offset points', # 坐标相对于数据点的坐标
             fontsize=16,       # 文本大小
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")) # 箭头

plt.plot([x, x],[0, np.cos(x)], 'b--', lw=2.5)  # 蓝色虚线
plt.scatter([x, ],[np.cos(x), ], s=50, c='blue')  # 该点处的cos值 蓝点

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$', # 显示文本
             xy=(x, np.cos(x)), 
             xycoords='data',
             xytext=(-90, -50), 
             textcoords='offset points', 
             fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# ----------------------------------------- 刻度值的修饰
for label in ax.get_xticklabels() + ax.get_yticklabels(): 
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='red', edgecolor='None', alpha=0.5))
# -----------------------------------------    

plt.show()
```
