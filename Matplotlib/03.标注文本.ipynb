# 03 标注文本

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### 文本标注示例

```python
fig = plt.figure(figsize=(12, 9))  # 指定图形的大小
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')  # 设置顶层标题

ax = fig.add_subplot(1, 1, 1)  # 插入一个子图
# fig.subplots_adjust(top=0.85)  # 这里好像不需要
ax.set_title('axes title')  # 子标题
ax.set_xlabel('xlabel')  # 设置x,y标签
ax.set_ylabel('ylabel')

ax.axis([0, 10, 0, 10])

ax.text(3, 8, 'boxed italics text in data coords', style='italic', bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
# 方框参数： 前景色、透明度、空白间距
ax.text(x=3, y=7, s='boxed italics text in data coords', fontdict={'color':'red', 'style':'italic'})
ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)  # 输入公式
ax.text(3, 2, 'unicode: Institut f\374r Festk\366rperphysik', fontsize=20)  # 拉丁文
ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',  # 垂直对齐，水平对齐
        transform=ax.transAxes,  # 转换轴 （这个不明白）
        color='green', fontsize=15)

ax.plot([2], [1], 'o')  # 绘制一个点 
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4), arrowprops=dict(facecolor='black', shrink=0.05))  # 添加注释

plt.show()
```

### 文本的对齐

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

left, width = .25, .5
bottom, height = .25, .5
right = left + width
top = bottom + height

fig = plt.figure(figsize=(10,7))
ax = fig.add_axes([0,0,1,1])

p = patches.Rectangle(  # patches.Rectangle(xy, width, height, angle=0.0, **kwargs)
    (left, bottom), width, height
    , fill=False  # 默认蓝色填充
    , transform=ax.transAxes  # 不知道什么作用
    , clip_on=False  # 不知道什么作用
    )
ax.add_patch(p)

# 调用方法 ax.text(x, y, s, fontdict=None, withdash=False, **kwargs)
# 这里省略了参数 transform=ax.transAxes

ax.text(left, bottom, '111111', ha='left', va='top', size='xx-large')

ax.text(left, bottom, '222222', ha='left', va='bottom', size='xx-large')

ax.text(right, top, '333333', ha='right', va='bottom', size='xx-large')

ax.text(right, top, '444444', ha='right', va='top', size='xx-large')

ax.text(right, bottom, '555555', ha='center', va='top', size='xx-large')

ax.text(left, 0.5*(bottom+top), '666666', ha='left', va='center', rotation='vertical', size='xx-large')

ax.text(0.5*(left+right), 0.5*(bottom+top), '777777', ha='center', va='center', fontsize=20, color='red')

ax.text(right, 0.5*(bottom+top), '888888', ha='center', va='center', rotation='vertical', size='xx-large')

ax.text(left, top, '999999', ha='center', va='center', rotation=45, size='xx-large')

ax.set_axis_off()
plt.show()
```

### 注释文本

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)  # 返回list,其实这里不需要名字来引用

ax.annotate('local max', xy=(2, 1), xytext=(2.5, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2,2)
plt.show()
```
