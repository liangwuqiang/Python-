# 02 使用style风格

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.style.available  # 查看可用的风格
```

```python
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.style.use('ggplot')  # R语言的ggplot风格
plt.plot(x, y)
```

```python
with plt.style.context(style=('dark_background')):  # 临时改变风格
    plt.plot(x, y, 'r-o')
```

```python
with plt.style.context(('dark_background')):
    pass
plt.plot(x, y, 'r-o')
```

```python
import matplotlib
matplotlib.get_configdir()  # 自定义风格的存放目录
```

内容以 属性：值 的形式保存：

axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 3
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16

假设我们将其保存为 mpl_configdir/stylelib/presentation.mplstyle，那么使用这个风格的时候只需要调用：

plt.style.use('presentation')
