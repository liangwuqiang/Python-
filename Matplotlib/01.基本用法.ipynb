### 控制显示的范围

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
# 指定 x 轴显示区域为 0-6，y 轴为 0-20
plt.axis([0,6,0,20])  # 参数顺序 x1, x2, y1, y2
```

### 传入Numpy数组

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')  # 三条曲线
```

```python
x = np.linspace(-np.pi, np.pi)  # np.linspace(start, stop, num=50)
y = np.sin(x)

plt.plot(x, y, linewidth=2.0, color='r')
```

```python
line, = plt.plot(x, y, 'r-')  # plt.plot(x, y, 'r-')返回的是list
line.set_antialiased(False)  # 将抗锯齿关闭   效果不明显 不知道有啥具体作用？
```

```python
lines = plt.plot(x, y)
plt.setp(lines, color='r', linewidth=2.0)  # 设置参数
```

### 子图

```python
def f(t):  # 定义函数
    return np.exp(-t) * np.cos(2*np.pi*t)
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t1
```

```python
t2 = np.arange(0.0, 5.0, 0.02)
t2
```

```python
plt.figure(1)
```

```python
plt.subplot(2, 1, 1)
```

```python
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')  # 蓝点，黑线
```

```python
plt.subplot(2, 1, 2)
```

```python
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')  # 红虚线
```

### 添加文字

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
```

```python
# plt.style.use('ggplot')
n, bins, patches = plt.hist(x=x, bins=50, normed=1, color='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  # 输入特殊的符号

plt.axis([40, 160, 0, 0.03])
plt.grid(True)
# plt.show()
```

```python
ax = plt.subplot(1, 1, 1)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(2.5, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))

plt.axis([-.3, 5.3, -2, 2])
# plt.ylim(-2,2)
```
